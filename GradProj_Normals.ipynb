```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread('IMG_3499.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Gaussian Blur to reduce noise and improve edge detection
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Use Canny edge detector with adjusted parameters
edges = cv2.Canny(gray, 50, 400)

# Adjust HoughLinesP parameters to better detect lines
lines = cv2.HoughLinesP(edges, rho=1, theta=np.pi/180, threshold=20, minLineLength=20, maxLineGap=5)

# Draw detected lines on the image
line_image = np.copy(image)
if lines is not None:
    for line in lines:
        x1, y1, x2, y2 = line[0]
        cv2.line(line_image, (x1, y1), (x2, y2), (255, 0, 0), 2)

# Display the result
plt.imshow(cv2.cvtColor(line_image, cv2.COLOR_BGR2RGB))
plt.title('Detected Lines')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import json

def compute_normals(lines, threshold_length=10, num_normals=5):
    normals = []
    for line in lines:
        x1, y1, x2, y2 = line[0]
        dx = x2 - x1
        dy = y2 - y1
        length = np.sqrt(dx**2 + dy**2)
        normal = (-dy / length, dx / length)  # Rotate 90 degrees to get the normal

        if length > threshold_length:
            for i in range(num_normals):
                t = i / (num_normals - 1)  # parameter t varies from 0 to 1
                mx = x1 + t * dx
                my = y1 + t * dy
                normals.append(((x1, y1, x2, y2), (mx, my), normal))
        else:
            mx, my = (x1 + x2) / 2, (y1 + y2) / 2  # Midpoint of the line segment
            normals.append(((x1, y1, x2, y2), (mx, my), normal))

    return normals

# Compute normals
normals = compute_normals(lines)

# Display normals
normal_image = np.copy(image)
for line, midpoint, normal in normals:
    x1, y1, x2, y2 = line
    mx, my = midpoint
    nx, ny = normal
    cv2.line(normal_image, (int(mx), int(my)), (int(mx + 20*nx), int(my + 20*ny)), (0, 255, 0), 2)

plt.imshow(cv2.cvtColor(normal_image, cv2.COLOR_BGR2RGB))
plt.title('Normals on Detected Lines')
plt.show()

# Prepare data to save in JSON
data_to_save = []
for (x1, y1, x2, y2), (mx, my), normal in normals:
    line_data = {
        'line': {'x1': int(x1), 'y1': int(y1), 'x2': int(x2), 'y2': int(y2)},
        'midpoint': {'mx': float(mx), 'my': float(my)},
        'normal': {'nx': float(normal[0]), 'ny': float(normal[1])}
    }
    data_to_save.append(line_data)

# Save the data to a JSON file
json_file_path = 'normals3NEW.json'
with open(json_file_path, 'w') as f:
    json.dump(data_to_save, f, indent=4)

files.download('normals3NEW.json')


print(f"Data saved to {json_file_path}")
```
